```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
```

```python
dados = np.array([
    [70.2, 12.5, 4.7],
    [65.1, 8.2, 3.9],
    [75.5, 15.6, 5.1],
    [80.3, 10.2, 4.5],
    [68.7, 11.8, 4.2],
    [72.9, 14.3, 5.3],
    [78.6, 9.8, 4.8],
    [66.4, 8.9, 4.0],
    [73.1, 13.7, 5.0],
    [69.5, 12.1, 4.3]
])
```

```python
kmeans = KMeans(n_clusters=3, random_state=0)
clusters = kmeans.fit_predict(dados)
```

```python
dados_com_clusters = np.column_stack((dados, clusters))
```

```python
for i in range(3):
    plt.figure()
    plt.title(f'Cluster {i+1}')
    plt.boxplot(dados_com_clusters[dados_com_clusters[:, -1] == i, :-1], vert=False)
    plt.xticks([1, 2, 3], ['Temperatura', 'Vibração', 'Corrente'])
    plt.show()
```
