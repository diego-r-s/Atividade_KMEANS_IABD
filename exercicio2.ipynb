```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

```

```python
setores = [1, 2, 3, 4, 5, 6, 7, 8]
produtos_fabricados = [100, 50, 15, 200, 500, 1000, 375, 450]

```

```python
df = pd.DataFrame({'Setores': setores, 'Produtos Fabricados': produtos_fabricados})

```

```python
from sklearn.preprocessing import StandardScaler

```

```python
scaler = StandardScaler()
df['Produtos Fabricados'] = scaler.fit_transform(df[['Produtos Fabricados']])

```

```python
kmeans = KMeans(n_clusters=3) 
kmeans.fit(df[['Setores', 'Produtos Fabricados']])
df['Cluster'] = kmeans.labels_

```

```python
plt.scatter(df['Setores'], df['Produtos Fabricados'], c=df['Cluster'], cmap='rainbow')
plt.xlabel('Setores')
plt.ylabel('Produtos Fabricados (normalizado)')
plt.title('Agrupamento de Setores')
plt.show()

```
